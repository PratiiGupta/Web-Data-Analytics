```python
import urllib.request
from bs4 import BeautifulSoup as bs
import pandas as pd
from datetime import datetime

def scrape_reviews(base_url, pages, brand):
    urls = [f'{base_url}?page={i+1}&sort=recency' for i in range(pages)]
    
    review_date = []
    reviewer = []
    rating = []
    review_text = []
    review_title = []
    brands = []

    headers = {'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'}

    for url in urls:
        req = urllib.request.Request(url, headers=headers)
        html = urllib.request.urlopen(req).read().decode("utf-8")
        soup = bs(html, 'html.parser')
        review_block = soup.find_all('article', class_='paper_paper__1PY90 paper_outline__lwsUX card_card__lQWDv card_noPadding__D8PcU styles_reviewCard__hcAvl')

        for i in review_block:
            try:
                name = i.find('span', class_='typography_heading-xxs__QKBS8 typography_appearance-default__AAY17').get_text()
                reviewer.append(name)
            except:
                reviewer.append("")

            try:
                title = i.find('h2', class_='typography_heading-s__f7029 typography_appearance-default__AAY17').get_text()
                review_title.append(title)
            except:
                review_title.append("")

            try:
                r_text = i.find('p', class_='typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn').get_text()
                review_text.append(r_text)
            except:
                review_text.append("")

            try:
                date = i.find('div', class_='typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_datesWrapper__RCEKH').get_text()
                review_date.append(datetime.strptime(date[-12:], "%b %d, %Y"))
            except:
                review_date.append("")

            try:
                rating_1 = i.find('div', class_='styles_reviewHeader__iU9Px')
                alt_text = rating_1.find('img', alt=True)
                rating.append(int(alt_text['alt'][6:7]))
            except:
                rating.append("")

            brands.append(brand)
                
    final_reviews = {
        'Review Date': review_date,
        'Reviewer': reviewer,
        'Review Rating': rating,
        'Review Title': review_title,
        'Review Text': review_text,
        'Brand': brands
    }

    return pd.DataFrame(final_reviews)

# Scrape reviews for each brand
df_oldnavy = scrape_reviews('https://www.trustpilot.com/review/oldnavy.ca', 20, 'OldNavy')
df_bananarepublic = scrape_reviews('https://www.trustpilot.com/review/bananarepublic.eu', 5, 'BananaRepublic')
df_gap = scrape_reviews('https://www.trustpilot.com/review/www.gap.com', 23, 'GAP')

# Combine dataframes
combined_df = pd.concat([df_oldnavy, df_bananarepublic, df_gap], ignore_index=True)

# Save the combined dataframe to a CSV
combined_df.to_csv('Combined_Reviews.csv', index=False)
```
